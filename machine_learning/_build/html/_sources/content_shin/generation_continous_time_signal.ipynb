## Continous Time Signal

### Experiment 1: Generation of sinusoidal signal

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Defining the independent variable
t = np.linspace(0, 1, 1000)

# Step 2: Defining the parameters of sine wave
A = 5  # Amplitude of sine wave
f = 5  # Frequency of sine wave
ph = 0 # Phase of sine wave

# Step 3: Expression of sine wave
x = A * np.sin(2 * np.pi * f * t + ph)

# Step 4: Finding the peak of the sine wave
# np.argmax(x) returns the indices of the maximum values along an axis.
peak_index = np.argmax(x)
peak_value = x[peak_index]
peak_time = t[peak_index]

# Step 5: Plotting the sine wave and marking the peak
plt.plot(t, x, label='Sine Wave')
plt.scatter(peak_time, peak_value, color='red', zorder=5, label='Peak', marker='+')  # Mark the peak
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title(r'A={}V, F={} Hz, $\phi={}^\circ$'.format(A, f, ph))
plt.legend()
plt.show()
```

### Experiment 1.2 Sinusoidal Signal with Different Phase

```python
# Step 1: Defining the independent variable
t = np.linspace(0, 1, 100)

# Step 2: Defining the parameters of sine wave
A = 1  # Amplitude
f = 5  # Frequency
phi = [0, 90, 180, 270]  # Phase angles in degrees

# Step 3: Creating subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Step 4: Generation and plotting of sine waves with different phases
for i, ax in enumerate(axes.flat):
    x = A * np.sin(2 * np.pi * f * t + phi[i] * np.pi / 180)
    ax.plot(t, x)
    ax.set_xlabel('Time (t)')
    ax.set_ylabel('Amplitude (V)')
    ax.set_title(r'$\Phi ={}^\circ$'.format(phi[i]))

# Step 5: Adjust layout
plt.tight_layout()
plt.show()
```

### Experiment 1.3 Generation of Three-Phase Sinusoidal Signal

```python
# Step 1: Defining the independent variable
t = np.linspace(0, 1, 100)

# Step 2: Defining the parameters of sine wave
A = 1  # Amplitude
f = 5  # Frequency

# Step 3: Defining the three different phases of sine wave
phi_1, phi_2, phi_3 = 0, 120, 240

# Step 4: Generation of sine waves for each phase
x1 = A * np.sin(2 * np.pi * f * t + phi_1 * np.pi / 180)
x2 = A * np.sin(2 * np.pi * f * t + phi_2 * np.pi / 180)
x3 = A * np.sin(2 * np.pi * f * t + phi_3 * np.pi / 180)

# Step 5: Plotting the result
plt.plot(t, x1, 'b', label='Phase-1')
plt.plot(t, x2, 'r', label='Phase-2')
plt.plot(t, x3, 'g', label='Phase-3')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Three phase sinusoidal signal')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

### Experiment 1.8 Generation of â€˜Sineâ€™ and â€˜Cosineâ€™ Functions from â€˜Complex Exponential Functionâ€™

```python
# Step 1: Generation of rotating phasor
t = np.linspace(0, 1, 100)
f = 5
x1 = np.exp(1j * 2 * np.pi * f * t)
x2 = np.exp(-1j * 2 * np.pi * f * t)

# Step 2: Generation of sine and cosine wave
x_cos = (x1 + x2) / 2
x_sin = (x1 - x2) / (2 * 1j)

# Step 3: Plotting the result
plt.subplot(2, 1, 1)
plt.plot(t, x_cos.real)  # Only the real part for cosine wave
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Cosine wave')

plt.subplot(2, 1, 2)
plt.plot(t, x_sin.real, 'r')  # Only the real part for sine wave
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Sine wave')

plt.tight_layout()
plt.show()
```

## Non-stationary Signal

### Experiment 1.11 Generation of Non-stationary Sinusoidal Signal

```python
# Frequencies for the signals
f1, f2, f3 = 0, 5, 10

# Time vectors for each segment
t1 = np.linspace(0, 1, 100)

# Generation of signal-1
x1 = np.sin(2 * np.pi * f2 * t1)

# Generation of signal-2
x2 = np.sin(2 * np.pi * f1 * t1)
x3 = np.sin(2 * np.pi * f3 * t1)

# Concatenation of signals
x = np.concatenate([x1, x2, x3])
y = np.concatenate([x3, x2, x1])

# Adjusting the time vector for concatenated signals
t = np.linspace(0, 3, 300)

# Plotting the result
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Signal-1')

plt.subplot(2, 1, 2)
plt.plot(t, y)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Signal-2')

plt.tight_layout()
plt.show()
```

### Experiment 1.12 Generation of Chirp Signal

The objective of this experiment is to generate chirp signal. The chirp signal can be considered as a frequency swept sinusoidal signal. Four different methods of frequency sweep are (1) linear, (2) quadratic, (3) logarithmic and (4) hyperbolic. In this experiment, the frequency sweep is from 10 Hz to 1 Hz, as considered

```python
# Generation of chirp signals
from scipy.signal import chirp

# Step 1: Generation of chirp signals
t = np.linspace(0, 10, 10000)
x1 = chirp(t, f0=10, f1=1, t1=10, method='linear')
x2 = chirp(t, f0=10, f1=1, t1=10, method='quadratic')
x3 = chirp(t, f0=10, f1=1, t1=10, method='logarithmic')
x4 = chirp(t, f0=10, f1=1, t1=10, method='hyperbolic')

# Step 2: Plotting the signals
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(t, x1)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Linear chirp')

plt.subplot(2, 2, 2)
plt.plot(t, x2)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Quadratic chirp')

plt.subplot(2, 2, 3)
plt.plot(t, x3)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Logarithmic chirp')

plt.subplot(2, 2, 4)
plt.plot(t, x4)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Hyperbolic chirp')

plt.tight_layout()
plt.show()
```

## Non-sinusoidal Waveform

### Experiment 1.13 Generation of Square Waveform

A square waveform is a non-sinusoidal periodic waveform. A square wave represents a sudden variation from â€˜ONâ€™ to â€˜OFFâ€™ state and vice versa. Duty cycle is the percentage of time a square wave remains high versus low over one period. Square waves are useful in modelling digital signals. Sine wave contains single frequency, whereas square wave contains a very wide bandwidth of frequencies.

```python
# Square wave with different duty cycles
from scipy import signal

# Step 1: Define the parameters
t = np.linspace(0, 1, 1000)  # Increased number of points for better resolution
f = 5
duty = [0.15, 0.25, 0.5, 0.75]

# Step 2: Generate and plot square waves with different duty cycles
plt.figure(figsize=(10, 8))

for i in range(len(duty)):
    x = signal.square(2 * np.pi * f * t, duty=duty[i])
    plt.subplot(2, 2, i + 1)
    plt.plot(t, x)
    plt.xlabel('Time (t)')
    plt.ylabel('Amplitude (V)')
    plt.ylim(-2, 2)
    plt.title('Square wave (duty={})'.format(duty[i]))

plt.tight_layout()
plt.show()
```

### Experiment 1.14 Generation of Sawtooth and Triangular Waveforms

Triangle and sawtooth waveforms are useful for exploring non-linearity in the circuit. A triangle waveform has uniform rise and fall time, whereas in a sawtooth waveform, the rise and fall times are markedly different.

```python
# Sawtooth wave with different widths

# Step 1: Define the parameters
t = np.linspace(0, 1, 1000)
f = 5
widths = [0.1, 0.2, 0.5, 0.8]

# Step 2: Generate and plot sawtooth waves with different widths
plt.figure(figsize=(10, 8))

for i, width in enumerate(widths):
    x = signal.sawtooth(2 * np.pi * f * t, width)
    plt.subplot(2, 2, i + 1)
    plt.plot(t, x)
    plt.xlabel('Time (t)')
    plt.ylabel('Amplitude (V)')
    plt.ylim(-2, 2)
    plt.title('Sawtooth wave (width={})'.format(width))

plt.tight_layout()
plt.show()
```

### Experiment 1.15 Generation of Sinc Function

A sinc function is an even function with a unit area. It is a symmetric function with respect to the origin. Fourier transform of sinc function will result in rectangular function and vice versa. Thus, sinc function is the impulse response of the ideal lowpass filter.

```python
# Generation of sinc function

# Step 1: Defining the independent variable
t = np.linspace(-10, 10, 1000)

# Step 2: Generating sinc function
x = np.sinc(t)

# Step 3: Plotting the sinc function
plt.figure(figsize=(8, 6))
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sinc function')
# plt.grid(True)
plt.tight_layout()
plt.show()
```

### Experiment 1.16 Generation of Rectangular and Triangular Pulse Signal

```python
# Generation of rectangular and triangular pulse signal

# Step 1: Generation of signals
t = np.linspace(-2, 2, 100)
rect_pulse = abs(t) < 1  # Rectangular pulse
tri_pulse = (1 - abs(t)) * (abs(t) < 1)  # Triangular pulse

# Step 2: Plotting of the pulse signals
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, rect_pulse)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Rectangular pulse')

plt.subplot(2, 1, 2)
plt.plot(t, tri_pulse)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude (V)')
plt.title('Triangular pulse')

plt.tight_layout()
plt.show()
```

### Task  

1. Write a python code to illustrate the fact that convolution of two rectangular pulse signals results in a triangular pulse.

```python
# Step 1: Define the time vector and the rectangular pulses
t = np.linspace(-2, 2, 100)
rect_pulse1 = np.where(np.abs(t) < 1, 1, 0)  # Rectangular pulse 1
rect_pulse2 = np.where(np.abs(t) < 1, 1, 0)  # Rectangular pulse 2

# Step 2: Convolution of the two rectangular pulses
conv_result = np.convolve(rect_pulse1, rect_pulse2, mode='full')

# Step 3: Define the time vector for the convolution result
t_conv = np.linspace(-4, 4, len(conv_result))

# Step 4: Plot the rectangular pulses and their convolution
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(t, rect_pulse1)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Rectangular Pulse 1')

plt.subplot(3, 1, 2)
plt.plot(t, rect_pulse2)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Rectangular Pulse 2')

plt.subplot(3, 1, 3)
plt.plot(t_conv, conv_result)
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.title('Convolution Result (Triangular Pulse)')

plt.tight_layout()
plt.show()
```

### Experiment 1.17 Generation of Gaussian Function

The Gaussian function is widely used in signal processing, image processing and communication fields. 

The expression for Gaussian function with the mean value $\mu$  and standard deviation $\sigma$ is given by

$$x(t) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(t - \mu)^2}{2\sigma^2}}.$$

```python
# Generation of Gaussian function

# Step 1: Defining the independent variable
t = np.linspace(-10, 10, 1000)
mu = 0  # Mean value
sigma = [0.01, 0.5, 1, 10]  # Standard deviations

# Step 2: Generate and plot Gaussian functions with different standard deviations
plt.figure(figsize=(10, 8))

for i in range(len(sigma)):
    k = 1 / (np.sqrt(2 * np.pi * sigma[i]))
    x = k * np.exp(-np.power(t - mu, 2.) / (2 * np.power(sigma[i], 2.)))
    plt.subplot(2, 2, i + 1)
    plt.plot(t, x)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title(r'$\sigma={} $'.format(sigma[i]))

plt.tight_layout()
plt.show()
```
